```python
import numpy as np
```

```python
INPUTFILE = 'input.txt'
```

```python
class Block:
    def __init__(self, lines):
        self.data = np.zeros((3,3))
        self.setbits = 0
        for i,line in enumerate(lines):
            for j, char in enumerate(line):
                if char == '#':
                    self.data[i,j] = 1.
                    self.setbits += 1
    def print(self):
        print(self.data)
```

```python
class Basket:
    def __init__(self):
        self.blocks = []
    def insert(self, lines):
        newblock = Block(lines)
        self.blocks.append(newblock)
    def print(self):
        for block in self.blocks:
            block.print()
    def isPossible(self, line):
        tokens = line.split()
        nums = np.array(tokens[1:],dtype=int)
        hw = tokens[0][:-1].split('x')
        h = int(hw[0])
        w = int(hw[1])
        numtotal = h*w
        numset = 0
        for i, num in enumerate(nums):
            numset += num*self.blocks[i].setbits
        return numtotal >= 1.1*numset
```

```python
basket = Basket()
activelines = []
valids = 0
with open(INPUTFILE, 'r') as file:
    for line in file:
        if 'x' in line:
            if basket.isPossible(line.strip()):
                valids+=1
        else:
            if '#' in line:
                activelines.append(line.strip())
        if len(activelines)==3:
            basket.insert(activelines)
            activelines = []
```
